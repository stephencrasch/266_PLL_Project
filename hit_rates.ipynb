```python
import pandas as pd 

predictions = pd.read_csv("predictions_r5.csv")
reviews_train = pd.read_csv("reviews_train.csv")
```

```python
joined = reviews_train.merge(predictions, on="review_id", how="left")
```

```python
joined
```

```python
pd.set_option('display.max_columns', None)
data = joined[['text', 'predicted_label','true_label','confidence']].sort_values(by='confidence', ascending=False)

data['correct'] = data['predicted_label'] == data['true_label']

# plot correct rate by confidence; higher confidence should correspond to higher accuracy
import matplotlib.pyplot as plt
data['confidence_bin'] = pd.cut(data['confidence'], bins=10)
accuracy_by_bin = data.groupby('confidence_bin')['correct'].mean()

plt.figure(figsize=(10, 6))
accuracy_by_bin.plot(kind='bar')
plt.title('Accuracy by Confidence Bin')
plt.xlabel('Confidence Bin')
plt.ylabel('Accuracy')
plt.xticks(rotation=45)
plt.show()

# now show grouped by predicted label
accuracy_by_label_bin = data.groupby(['predicted_label', 'confidence_bin'])['correct'].mean().unstack()
accuracy_by_label_bin
```
